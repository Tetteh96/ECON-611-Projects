# Project Title: Quantifying the Impact of Trade Agreements on Economic Growth (with the United States, Canada and Mexico as a Case Study)
## Introduction:
Many government officials in both developed and developing countries believe that trade liberalization, including Regional Trade Agreements (RTAs), will lead to faster economic growth, which is a key goal. However, there are some who are skeptical of the positive impact of RTAs on regional development (Lloyd and Paun, 2018). Despite the fact that economic growth is a major motivation for regional integration, its actual effects have not yet been observed. For example, Zambia after liberalizing its economy in the 1990s, experienced high levels of debt, inflation, and unemployment. While some sectors, such as mining and agriculture, grew rapidly, overall economic growth was slow. On the other hand, East Asian nations that have joined RTAs at a later stage have experienced more rapid economic growth compared to the global average. As a result, studying the impact of RTAs on economic growth is highly pertinent to the ongoing discussion on regionalism (Park and Lee, 2019).

The United States, Canada, and Mexico have been in a trilateral trade agreement (an RTA), known as the North American Free Trade Agreement (NAFTA) for more than 27 years (Office of the United States Trade Representative, n.d.). This agreement has been a subject of numerous debates and discussions regarding its impacts on the economic growth of these countries. NAFTA was in effect from 1994 until the United States-Mexico-Canada Agreement (USMCA) replaced it on July 1, 2020. Therefore, this study aims to investigate the quantitative impact of the NAFTA trade agreement on the economic growth of these countries.
## Research Question:
What is the quantitative impact of trade agreement (NAFTA) on economic growth in the United States, Canada, and Mexico?
## Methodology:
The study will use a panel data analysis technique, which allows for the examination of a large dataset of time series data (Hsiao, 2014). The data set used for this analysis will be from the World Bank, ILO, IMF and OECD, and it will consist of the annual GDP growth rate for the three countries, covering the period of 1960 to 2021. The study will employ an OLS regression to examine the impact of NAFTA on economic growth, with the dependent variable being the annual growth rate of GDP in each country. The independent variable would be the trade agreement, measured as a binary variable, and the control variables would include population growth, inflation rate, trade openness, employment rate, and net FDI.

_Variables_:

_Dependent Variable_: The dependent variable is economic growth, measured as the annual growth rate of GDP in each country.

_Independent Variable_: The primary independent variable is the trade agreement, measured as a binary variable (1 if NAFTA is in effect, and 0 otherwise).

_Control Variables_: The control variables include population growth, inflation rate, employment, net FDI and trade openness.

* To measure the variable "trade openess," the study will use the World Bank's "Trade Openness Index," which is a composite index that captures the extent to which a country's economy is open to trade. The index is constructed as the sum of exports and imports of goods and services as a share of GDP.

* To measure the variable "inflation," the study will use the Consumer Price Index (CPI), which is a commonly used measure of inflation. The CPI measures the changes in the prices of a basket of goods and services that are typically consumed by households.

* To measure the variable "population growth," the study will use the annual population growth rate, which is the percentage change in a country's population over a year. This data can be obtained from the World Bank.

* Employment rate is an important economic indicator that measures the percentage of the labor force that is employed, and it could be an important factor in determining the impact of trade agreements on economic growth.

* Net FDI, or foreign direct investment, is also an important factor in economic growth, as it represents investment from foreign entities into a country's economy. These variables in the analysis could help to provide a more comprehensive understanding of the impact of trade agreements on economic growth.


The results of this analysis would allow me to assess the impact of the trade agreements on economic growth, while controlling for the effects of these additional variables. It could provide insights into the relationship between trade agreements, employment, inflation, population growth, trade openess, and foreign investment, and help policymakers understand how these factors contribute to economic growth in the United States, Canada, and Mexico.

### Handeling Missing Values

```python
# create imputer object with mean strategy
imputer = SimpleImputer(strategy='mean')

# impute missing values in the dataframe
trade_df_imputed = pd.DataFrame(imputer.fit_transform(trade_df),
                                 columns=trade_df.columns)

# separate input and output variables
X = trade_df_imputed.drop(['TRADE_OPEN', 'GDP_GROWTH', 'EMPLOYMENT', 'NET_FDI'], axis=1)
y = trade_df_imputed[['TRADE_OPEN', 'GDP_GROWTH', 'EMPLOYMENT', 'NET_FDI']]

# Train two separate models, one for each target variable
model_trade = LinearRegression()
model_trade.fit(X, y['TRADE_OPEN'])

model_gdp = LinearRegression()
model_gdp.fit(X, y['GDP_GROWTH'])

model_emp = LinearRegression()
model_emp.fit(X, y['EMPLOYMENT'])

model_fdi = LinearRegression()
model_fdi.fit(X, y['NET_FDI'])

# Predict missing values
gdp_pred = model_gdp.predict(X)
emp_pred = model_emp.predict(X)

# Replace missing values in the original dataframe with predicted values
trade_df.loc[trade_df['TRADE_OPEN'].isnull(), 'TRADE_OPEN'] = emp_pred[trade_df['TRADE_OPEN'].isnull()]
trade_df.loc[trade_df['GDP_GROWTH'].isnull(), 'GDP_GROWTH'] = gdp_pred[trade_df['GDP_GROWTH'].isnull()]
trade_df.loc[trade_df['EMPLOYMENT'].isnull(), 'EMPLOYMENT'] = emp_pred[trade_df['EMPLOYMENT'].isnull()]
trade_df.loc[trade_df['NET_FDI'].isnull(), 'NET_FDI'] = emp_pred[trade_df['NET_FDI'].isnull()]
```

### CUSUM (Cummulative Sum) Test to check Structural Break

```python
# Fit a linear regression model to the entire data set
model = sm.OLS(trade_df['log_TRADE_OPEN'], sm.add_constant(trade_df.drop('log_TRADE_OPEN', axis=1))).fit()

# Perform the CUSUM test
cusum_test = sm.stats.diagnostic.breaks_cusumolsresid(model.resid, ddof=1)

# Print the results
print(cusum_test)
```

The CUSUM test statistic is 2.086 and the p-value is 0.0003. This indicates that there is strong evidence of a structural break in the data, as the probability of obtaining such an extreme test statistic under the null hypothesis is very low.

The list of critical values provides a reference for interpreting the significance of the test result at different levels. The list includes critical values for three levels of significance (0.1, 0.05, and 0.01) and for different sample sizes (1, 5, and 10). The critical values increase with the sample size, reflecting the increased likelihood of detecting a structural break with more data.

Overall, the CUSUM test suggests that there is a significant change in the slope of the regression line over time, indicating a structural break in the data. This finding has important implications for the validity of the regression analysis and the interpretation of the coefficients. It may suggest the need for a more complex model that accounts for the changing relationship between the dependent and independent variables over time.

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan


# Define the response variable and predictor variables
y = trade_df['log_TRADE_OPEN']
X = trade_df.drop(columns= 'log_TRADE_OPEN')

# Fit a linear regression model
model = sm.OLS(y, sm.add_constant(X)).fit()

# Compute the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)

# Print the test results
print('Breusch-Pagan test results:')
print('Test statistic:', bp_test[0])
print('p-value:', bp_test[1])
print('F-statistic:', bp_test[2])
print('F p-value:', bp_test[3])
```

The test statistic is 82.57300885942695 and the p-value is 4.973333908606412e-14, which is very small. This indicates strong evidence against the null hypothesis of homoscedasticity, suggesting the presence of heteroskedasticity in the model.

The F-statistic is 17.959321670113813, with a corresponding p-value of 2.1822956658649534e-19. This suggests that the heteroskedasticity is statistically significant, further supporting the conclusion of the Breusch-Pagan test.

The results of the Breusch-Pagan test indicate that there is strong evidence of heteroskedasticity in the regression model. This should be taken into account when interpreting the results of the regression analysis and appropriate measures should be taken to address this issue, such as using a weighted regression or transforming the data.

```python
from statsmodels.tsa.stattools import adfuller

# Define a function to perform the ADF test and print the results
def adf_test(timeseries):
    # Perform the ADF test
    result = adfuller(timeseries)
    # Extract and print the test statistic and p-value
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    # Print the critical values at various confidence levels
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

# Call the function with the time series data
adf_test(trade_df['log_TRADE_OPEN'])
```

Based on the Augmented Dickey-Fuller (ADF) test results, the ADF statistic of -1.467 is greater than the critical values of -3.467, -2.878, and -2.575 at the 1%, 5%, and 10% significance levels, respectively. Additionally, the p-value of 0.55 is higher than the significance level of 0.05.

This suggests that we fail to reject the null hypothesis of the ADF test, which is that the time series has a unit root and is non-stationary. Therefore, based on these results, we can conclude that the time series is likely non-stationary.

Since the data is not stationary, we need to transform the data by differencing it to make it stationary.

```python
from scipy.stats import chi2

# calculate the predicted values
y_pred = results.predict(X)

# calculate the residuals
resid = y - y_pred

# perform the chi-square goodness-of-fit test
n = len(resid)
k = 10  # number of independent variables
p = k + 1  # number of parameters (including the intercept)
df = n - p
chi2_stat = n * np.sum(resid ** 2) / results.mse_resid
p_value = 1 - chi2.cdf(chi2_stat, df)

print('Chi-square statistic:', chi2_stat)
print('Degrees of freedom:', df)
print('p-value:', p_value)
```
