## use a for loop to call the first index in the two_array as i

### calling two_dim_array[i] will yield the $i^{th}$ row

```python
# Loop through the rows of the two-dimensional array
for i in range(len(two_dim_array)):
    # Print the row index
    print(i)
    # Print the current row
    print(two_dim_array[i])
```

### passing the ros to a second for loop will cycle through
### the second (row) index as $j$. Calling row [$j$] will yield 
### the $i^{th}$ value from the row

```python
# Print the header for the output
print('i j')

# Loop through the rows of the two-dimensional array
for i in range(len(two_dim_array)):
    # Get the current row
    row = two_dim_array[i]
    
    # Loop through the elements in the current row
    for j in range(len(row)):
        # Print the row and column indices
        print(i, j)
        # Print the value at the current position
        print('val:', row[j])
```

### passing the row to a second for loop will cycle through the second (row) 

```python
# Loop through the rows of the two-dimensional array
for row in two_dim_array:
    # Print the current row
    print(row)
    # Loop through the values in the current row
    for val in row:
        # Print the current value
        print(val)
```

```python
# create article date
# Create a dictionary of dictionaries to store macroeconomic data
macro_dict = {'GDP' : {},
            'Money' : {},
            'Real GDP' : {},
            'Price Level' : {}}

# Loop through each key in the dictionary
for key in macro_dict.keys():
    # Loop through each year from 1990 to 2020
    for i in range(1990, 2021):
        # Assign a random value to the key-year combination
        macro_dict[key][i] = np.random.random() * 10000

# call the dictionary
macro_dict



```

```python
macro_DF = pd.DataFrame(macro_dict)
macro_DF
```

```python
macro_DF[['GDP']]
```

```python
# cerate a new variable 'Velocity'
macro_DF['Velocity'] = macro_DF['Price Level'].mul(macro_DF['Real GDP'].div(macro_DF['Money']))
macro_DF
```

```python
# plot using method matplotlib from the pandas DataFrame

# Plotting macroeconomic data over time

# Initialize the plot with a specified figure size
fig, ax = plt.subplots(figsize = (20, 12))

# Plot the macroeconomic data in the macro_DF dataframe, specifying line width and transparency
macro_DF.plot(linewidth= 5, alpha= .7, ax = ax)

# Add a legend to the plot, with font size 20
ax.legend(fontsize= 20)
```

```python
# loop through the columns (keys) of the macro_DF dataframe
for key in macro_DF:
    # call the values of each column
    macro_DF[key]
```

```python
# create a multiplot visualisation with a separate plot area for each variables
# count the number of variables
num_vars = len(macro_DF.keys())

# create a plot with a subplot for each variable (total 5)
fig, ax = plt.subplots(1, num_vars, figsize= (21, 6))

# to plot each variable with on each plot, cycle through index of the keys
for i in range(num_vars):
    # call key by index number
    key = list(macro_DF.keys())[i]
    # set color in light of number i so that it is 'Ci'
    color = 'C' + str(i)
    alpha = .1 + .13 * i
    # plot key according to index number in ax
    macro_DF[key].plot.line(color = color, alpha = alpha, ax = ax[i])
    ax[i].set_title(key, fontsize=20)
```

## Plotting Income Distribution

```python
income_data = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])
income_data = income_data.rename(columns = {"Unnamed: 0": "Household Income"})
income_data
```

```python
# Selecting rows 1 to the second-to-last row of the dataframe 'income_data' and assign it to a new variable 'income_data_filtered'
#income_data.iloc[1 : -1]
```

```python
# save series of persons in each income level   
#Select the 'Number' column excluding the first and last rows
income = income_data['Number'][1 : -1].div(1000)

#Set the first element of the 'income' series to 0
income.iloc[0] = 0

#Display the first 5 elements of the 'income' series
income.head()
```

```python
# make a list that will serve as the index to replace current index

# Define a list of income levels with increments of 5 from 0 to 200, and with an additional value of 250
income_levels = list(range(0, 201, 5))

# Append the value 250 to the list
income_levels.append(250)

# Convert each value in the list to a string with "k" appended to the end, representing thousands of dollars
income_levels = [str(val) + 'k' for val in income_levels]
income_levels.append('250k+')
income_levels
```

```python
# accessing the 'Household Income' column of the income_data DataFrame. 
income_data['Household Income']
```

```python
income_data = pd.DataFrame({"Income Level":np.array(income_levels),
                           "Households in Bracket":income})
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()
income_data.set_index("Income Level")
income_data.head()
```

```python
total_households = income_data['Households in Bracket'].sum().round(3)
income_data['% in Bracket'] = income_data['Households in Braacket'].div(total_households)
income_data['Comulative'] = income_data['% in Bracket'].cusum()
```
