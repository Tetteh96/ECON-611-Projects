## use a for loop to call the first index in the two_array as i

### calling two_dim_array[i] will yield the $i^{th}$ row

```python
# Loop through the rows of the two-dimensional array
for i in range(len(two_dim_array)):
    # Print the row index
    print(i)
    # Print the current row
    print(two_dim_array[i])
```

### passing the ros to a second for loop will cycle through
### the second (row) index as $j$. Calling row [$j$] will yield 
### the $i^{th}$ value from the row

```python
# Print the header for the output
print('i j')

# Loop through the rows of the two-dimensional array
for i in range(len(two_dim_array)):
    # Get the current row
    row = two_dim_array[i]
    
    # Loop through the elements in the current row
    for j in range(len(row)):
        # Print the row and column indices
        print(i, j)
        # Print the value at the current position
        print('val:', row[j])
```

### passing the row to a second for loop will cycle through the second (row) 

```python
# Loop through the rows of the two-dimensional array
for row in two_dim_array:
    # Print the current row
    print(row)
    # Loop through the values in the current row
    for val in row:
        # Print the current value
        print(val)
```

## Plotting Income Distribution

```python
# Read the Excel file from the URL into a pandas dataframe
income_data = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])

# Rename the first column to 'Household Income'
income_data = income_data.rename(columns = {"Unnamed: 0": "Household Income"})

# Display the dataframe
income_data
```

```python
# Selecting rows 1 to the second-to-last row of the dataframe 'income_data' and assign it to a new variable 'income_data_filtered'
income_data.iloc[1 : -1]
```

```python
# save series of persons in each income level   
#Select the 'Number' column excluding the first and last rows
income = income_data['Number'][1 : -1].div(1000)

#Set the first element of the 'income' series to 0
income.iloc[0] = 0

#Display the first 5 elements of the 'income' series
income.head()
```

```python
# make a list that will serve as the index to replace current index

# Define a list of income levels with increments of 5 from 0 to 200, and with an additional value of 250
income_levels = list(range(0, 201, 5))

# Append the value 250 to the list
income_levels.append(250)

# Convert each value in the list to a string with "k" appended to the end, representing thousands of dollars
income_levels = [str(val) + 'k' for val in income_levels]
income_levels.append('250k+')
income_levels
```

```python
# accessing the 'Household Income' column of the income_data DataFrame. 
income_data['Household Income']
```

```python
# Selecting two arrays of data and pass them into a dataframe using dictionary function

# Create new columns 'Income Level' and 'Households in Bracket'
income_data = pd.DataFrame({"Income Level": np.array(income_levels),
                           "Households in Bracket": income})

# Calculate the cumulative sum of the 'Households in Bracket' column and add it as a new column 'Cumulative'
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()

# Set the index of the dataframe to 'Income Level'
income_data.set_index("Income Level", inplace=True)

# Display the first 5 rows of the dataframe
income_data.head()
#income_data.plot.line(secondary_y= 'Cumulative')
```

```python
# Calculate the total number of households
total_households = income_data['Households in Bracket'].sum().round(3)

# Calculate the percentage of households in each income bracket
income_data['% in Bracket'] = (income_data['Households in Bracket'] / total_households).mul(100).round(3)

# Calculate the cumulative percentage of households in each income bracket
income_data['Cumulative %'] = income_data['% in Bracket'].cumsum().round(3)

# Display the updated dataframe
income_data
```

```python
# rcparams allows to access global variables in matplotlib and 
# then set a default value for them

# Update the font size in the plotting parameters
plt.rcParams.update({'font.size' : 16}) 

# Create a dataframe with only the columns '% in Bracket' and 'Cumulative %'
plot_df = income_data[['% in Bracket', 'Cumulative %']]

# Create a figure and axis object for plotting
fig, ax = plt.subplots(figsize= (15, 7))

# Plot the data in the plot_df dataframe as a bar plot
# Use a secondary y-axis for the cumulative percentage
plot_df.plot.bar(secondary_y= 'Cumulative %', width = .90, alpha = .75, color= ['C1', 'C2'], ax = ax)

# Format the x-axis tick labels to include dollar signs
x_ticklabels = ['$' + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

# Format the y-axis tick labels to include percent signs
y1_ticklabels = [str(y) + '%' for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)

# Show the plot
plt.show()

```

```python
# build a distribution of income with 25k bins instead of 5k bins
# start by counting through the rows by 5s in the original dataframe
income_25k_bin = income_data.loc[ ::5]

# we were missing the final row with 250k+, so add this in the last row
# we created a new row using df.loc[new_row_name]
income_25k_bin.loc['250k+'] = income_data.iloc[-1]

# redifine the marginal bins by taking the difference of the cumulative bins
income_25k_bin['Households in Bracket'] = income_25k_bin['Cumulative']
income_25k_bin['Households in Bracket'] = income_25k_bin['Households in Bracket'].diff().fillna(0)

# the information from the % in Bracket column was incorrect. So we recalculate this using
# the new Households in Bracket data
income_25k_bin['% in Bracket'] = income_25k_bin['Households in Bracket'].div(total_households).mul(100).round(3)
income_25k_bin
```

```python
# Update the font size in the plotting parameters
plt.rcParams.update({'font.size' : 20})

# Create a dataframe with only the columns '% in Bracket' and 'Cumulative %'
# Exclude the first row
plot_df = income_25k_bin[['% in Bracket', 'Cumulative %']].iloc[1:]

# Create a figure and axis object for plotting
fig, ax = plt.subplots(figsize= (15, 10))

# Plot the data in the plot_df dataframe as a bar plot
# Use a secondary y-axis for the cumulative percentage
plot_df.plot.bar(secondary_y= 'Cumulative %', width = .99, alpha = .75, color= ['C3', 'C2'], ax = ax)

# Format the x-axis tick labels to include dollar signs
x_ticklabels = ['$' + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

# Format the y-axis tick labels to include percent signs
y1_ticklabels = [str(y) + '%' for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)

# Set the title of the plot
ax.set_title("Income Distribution in the United States")

# Show the plot
plt.show()
```
