```python
# Import a dataset and print the dataframe
macro_data = pd.read_csv(r'C:\Users\akoto\OneDrive\Documents\GitHub\ECON-611-Projects\Ch3\macro_df.csv')
macro_data
```

```python
# Create a new column of data derived from the data that was imported
macro_data['Velocity_Money_Ratio'] = macro_data['Velocity'] / macro_data['Money']
macro_data
```

```python
# Use a for loop to print individually each column in the dataframe
for column in macro_data.columns:
    print(macro_data[column])
```

```python
# Create a function that plots 2-3 different variables from the dataframe
def plot_data(df, names,):
    fig, ax = plt.subplots(figsize = (10, 8))
    for name in names:
        ax.plot(df[name], label=name)
    ax.set_xlabel('')
    ax.set_ylabel('Value')
    ax.set_title('Macro Economic Indicators')
    ax.legend()
    plt.show()

plot_data(macro_data, ['Real GDP', 'Velocity'])
```

```python
# Create a bar graph plot using the data that was imported
fig, ax = plt.subplots(figsize= (15, 10))
macro_data.head(5).plot(kind='bar', ax=ax) # ploting first 5 rows of the dataset
ax.set_title('Macroeconomic Indicators')
ax.set_xlabel('')
ax.set_ylabel('Indicator Value')
plt.show()
```

```python
# Create a bar graph plot using the data that was imported
fig, ax = plt.subplots(figsize= (15, 10))
macro_data.head(5).plot(kind='bar', edgecolor='black', linewidth=2, ax=ax) # ploting first 5 rows of the dataset
ax.set_title('Macroeconomic Indicators', fontsize= 20)
ax.set_xlabel('')
ax.set_ylabel('Indicator Value', fontsize= 14)
ax.tick_params(labelsize=12)
ax.set_xticks([])
plt.show()
```
